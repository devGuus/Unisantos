```python
# Pacote de processamento de dados
import pandas as pd
import numpy as np

# Pacote de visualização de dados
import matplotlib.pyplot as plt
```

```python
# Leitura do arquivo

arq = 'https://github.com/devGuus/Unisantos/raw/main/tde/Dados/importacoes-exportacoes-petroleo-2000-2023.xls'
arquivo1 = pd.read_excel(arq)

# Imprimindo arquivo completo

arquivo1
```

```python
# Para obter as (n) primeiras linhas do arquivo usamos head(n)

arquivo1.head(5)
```

```python
# Para obter as (n) ultimas linhas do arquivo usamos tail(n)

arquivo1.tail(5)
```

```python
# Imprime as informações gerais do arquivo:

arquivo1.info()
```

```python
# Descreve algumas estatísticas basiscas do arquivo:

arquivo1.describe() 
```

```python
# Mostra o número de valores únicos para cada coluna:

arquivo1.nunique()
```

```python
# Colocando o mês como primeira coluna e somando o numero de linhas e colunas
# Para usar uma coluna expecifica podemos usar (...).sum()['nomeDaColuna'])

pd.DataFrame(arquivo1.groupby(["MÊS", "ANO"]).sum())
```

```python
# Verifica os valores nulos de cada coluna.

print('Colunas:')
print(arquivo1.isnull().sum())
print()

# Número de linhas e colunas

num_lin, num_cols = arquivo1.shape
print(f"O conjunto de dados tem {num_lin} linhas e {num_cols} colunas.")
print()

# Nomes de cada coluna

column_names =arquivo1.columns.tolist()
print(f"Os nomes das colunas são: {column_names}")
print()

# Obter a média, mediana e devio do padrão

med_valor = arquivo1['IMPORTADO / EXPORTADO'].mean()
mediana_valor = arquivo1['IMPORTADO / EXPORTADO'].median()
desvio = arquivo1['IMPORTADO / EXPORTADO'].std()
print('A média é {}, a mediana é {} e o desvio de padrão é {}' .format(med_valor, mediana_valor, desvio))

# Função que agrupa os dados em uma coluna para obter estatísticas 

group_data = arquivo1.groupby('IMPORTADO / EXPORTADO').describe()
print(group_data)
```

```python
# Pegando informação expecifica do arquivo

df_export = arquivo1[(arquivo1['OPERAÇÃO COMERCIAL'] == 'EXPORTAÇÃO') & (arquivo1['ANO'] == 2022)]

# Plota o gráfico de barras
plt.bar(df_export['MÊS'], df_export['DISPÊNDIO / RECEITA'])
plt.xlabel('Mês')
plt.ylabel('Quantidade Exportada (1000 m³)')
plt.title('Exportação no ano 2022')
plt.show()
```

```python
# seleciona colunas para analise
colunas_analise = ['OPERAÇÃO COMERCIAL', 'IMPORTADO / EXPORTADO', 'DISPÊNDIO / RECEITA']
arquivo1_selecionado = arquivo1[colunas_analise]

# Exibir o DataFrame resultante
print(arquivo1_selecionado)
```

```python
# filtra as linhas em que a coluna "OPERAÇÃO COMERCIAL" contém "IMPORTAÇÃO"
arquivo1_importacao = arquivo1[(arquivo1['OPERAÇÃO COMERCIAL'] == 'IMPORTAÇÃO')]

# filtra as linhas em que a coluna "OPERAÇÃO COMERCIAL" contém "EXPORTAÇÃO"
arquivo1_exportacao = arquivo1[(arquivo1['OPERAÇÃO COMERCIAL'] == 'EXPORTAÇÃO')]

# calcula o total de gastos para importação e exportação
total_importacao = arquivo1_importacao['DISPÊNDIO / RECEITA'].sum()
total_exportacao = arquivo1_exportacao['DISPÊNDIO / RECEITA'].sum()

# cria uma lista com os valores de gastos
valores = [total_importacao, total_exportacao]

# cria uma lista com os rótulos
rotulos = ['Importação', 'Exportação']

# plotar o gráfico de pizza
plt.pie(valores, labels=rotulos, autopct='%1.1f%%', startangle=90)

# adiciona título ao gráfico
plt.title('Proporção de Gastos entre Importação e Exportação')

# exibe o gráfico
plt.show()
```

```python
# filtra as linhas em que OPERAÇÃO COMERCIAL é igual a IMPORTAÇÃO
arquivo1_importacao = arquivo1[arquivo1['OPERAÇÃO COMERCIAL'] == 'IMPORTAÇÃO']

# renomeia as colunas
arquivo1_importacao = arquivo1_importacao.rename(columns={'IMPORTADO / EXPORTADO': 'IMPORTAÇÃO', 'DISPÊNDIO / RECEITA': 'DISPÊNDIO'})

# remove a coluna OPERAÇÃO COMERCIAL
arquivo1_importacao = arquivo1_importacao.drop(columns='OPERAÇÃO COMERCIAL')

# remove o ano 2023
arquivo1_filtered = arquivo1_importacao[arquivo1_importacao['ANO'] < 2023]

# agrupa os dados por ano e soma o dispendio
dispendio_por_ano = arquivo1_filtered.groupby('ANO')['DISPÊNDIO'].sum()

# encontra o índice da maior e menor dispendio
maior_dispendio = dispendio_por_ano.argmax()
menor_dispendio = dispendio_por_ano.argmin()

# define as cores das barras
cores = ['r' if i == maior_dispendio else 'g' if i == menor_dispendio else 'b' for i in range(len(dispendio_por_ano))]

# plota o gráfico de barras
dispendio_por_ano.plot(kind='bar', color=cores, figsize=(10,5))

# define os títulos do gráfico e dos eixos
plt.title('Dispêndio por ano (com base nas Importações)')
plt.xlabel('Ano')
plt.ylabel('Dispêndio')

# exibe o gráfico
plt.show()
```

```python
# filtra as linhas em que OPERAÇÃO COMERCIAL é igual a IMPORTAÇÃO
arquivo1_exportacao = arquivo1[arquivo1['OPERAÇÃO COMERCIAL'] == 'EXPORTAÇÃO']

# renomeia as colunas
arquivo1_exportacao = arquivo1_exportacao.rename(columns={'IMPORTADO / EXPORTADO': 'EXPORTAÇÃO', 'DISPÊNDIO / RECEITA': 'RECEITA'})

# remove a coluna OPERAÇÃO COMERCIAL
arquivo1_exportacao = arquivo1_exportacao.drop(columns='OPERAÇÃO COMERCIAL')

# remove o ano 2023
arquivo1_filtered = arquivo1_exportacao[arquivo1_exportacao['ANO'] < 2023]

# agrupa os dados por ano e soma o dispendio
receita_por_ano = arquivo1_filtered.groupby('ANO')['RECEITA'].sum()

# encontra o índice da maior e menor dispendio
maior_receita = receita_por_ano.argmax()
menor_receita = receita_por_ano.argmin()

# define as cores das barras
cores = ['r' if i == menor_receita else 'g' if i == maior_receita else 'b' for i in range(len(receita_por_ano))]

# plota o gráfico de barras
receita_por_ano.plot(kind='bar', color=cores, figsize=(10,5))

# define os títulos do gráfico e dos eixos
plt.title('Receita por ano (com base nas Exportações)')
plt.xlabel('Ano')
plt.ylabel('Receita')

# exibe o gráfico
plt.show()
```
